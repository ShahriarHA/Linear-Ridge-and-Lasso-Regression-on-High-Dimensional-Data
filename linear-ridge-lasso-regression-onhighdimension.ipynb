# Variance Inflation Factor(VIF) #   

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
x_const = add_constant(x)
vif = pd.DataFrame()
vif["Feature"] = x_const.columns
vif["VIF"] = [variance_inflation_factor(x_const.values, i) for i in range(x_const.shape[1])]
print(vif.sort_values("VIF", ascending=False).head(20)) 
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error
```

# Linear Regression

```python
LR_model = LinearRegression()
```

```python
LR_model.fit(xTrain,yTrain)
```

```python
yPred = LR_model.predict(xTest)
yPred
```

```python
trainingAccuracy = LR_model.score(xTrain,yTrain)*100
testAccuracy = LR_model.score(xTest,yTest)*100
print(trainingAccuracy,testAccuracy)
```

```python
LR_mse = mean_squared_error(yTest,yPred)
LR_mse
```

# L1 Regularization-Lasso

```python
l1_model = Lasso()
```

```python
l1_model.fit(xTrain,yTrain)
```

```python
lassoPrediction = l1_model.predict(xTest)
lassoTestAccuracy = l1_model.score(xTest,yTest)*100
lassoTrainingAccuracy = l1_model.score(xTrain,yTrain)*100
lassoMSE = mean_squared_error(yTest,lassoPrediction)
```

```python
print(lassoTrainingAccuracy,lassoTestAccuracy)
print(lassoMSE)
```

# L2 Regularization-Ridge

```python
l2_model = Ridge()
l2_model.fit(xTrain,yTrain)
```

```python
ridgePrediction = l2_model.predict(xTest)
ridgeMSE = mean_squared_error(yTest,ridgePrediction)
ridgeTestScore = l2_model.score(xTest,yTest)*100
ridgeTrainingScore = l2_model.score(xTrain,yTrain)*100
```

```python
print(ridgeTrainingScore,ridgeTestScore)
print(ridgeMSE)
```

## Changing alpha values on ridge regression

```python
alphas = [5,10,50,60,100]
for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(xTrain,yTrain)
    print(f"when Alpha = {a}, The Training Score is: {ridge.score(xTrain,yTrain)*100:.4f}, The test score is {ridge.score(xTest,yTest)*100:.4f}")
    
```

## Changing alpha values on Lasso regression

```python
alphas = [5,10,50,60,100,1000,10000]
for a in alphas:
    lasso = Lasso(alpha=a)
    lasso.fit(xTrain,yTrain)
    print(f"when Alpha = {a}, The Training Score is: {lasso.score(xTrain,yTrain)*100:.4f}, The test score is {lasso.score(xTest,yTest)*100:.4f}")
    
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
xTrain_scaled = scaler.fit_transform(xTrain)
xTest_scaled = scaler.transform(xTest)

scaledRidge = Ridge()
scaledRidge.fit(xTrain_scaled,yTrain)
sRidgeTrain = scaledRidge.score(xTrain_scaled,yTrain)
SRidgeTest = scaledRidge.score(xTest_scaled,yTest)#test score
```

```python
print(sRidgeTrain,SRidgeTest)
```

### droping high VIF features

```python
high_vif_features = vif[vif["VIF"] > 1000]["Feature"].tolist()
x_reduced = x.drop(columns=high_vif_features)
```

```python
x_reduced
```

```python
xReTrain,xReTest,yTrain,yTest = train_test_split(x_reduced,y,test_size = 0.20, random_state=42)
```

```python
lr_model_v1 = LinearRegression()
lr_model_v1.fit(xReTrain,yTrain)
print(lr_model_v1.score(xReTrain,yTrain)*100)#training score
print(lr_model_v1.score(xReTest,yTest)*100)#test score
```

*Most of the features in our dataset are highly correlated as we can see from **VIF code section**, which makes Linear Regression unstable and gives poor results. That is why Ridge and Lasso performed slightly better for this high-dimensional dataset.*
